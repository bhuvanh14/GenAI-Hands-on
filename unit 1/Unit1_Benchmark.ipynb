```python
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")
```

```python
prompt = "The future of Artificial Intelligence is"

bert_gen = pipeline("text-generation", model="bert-base-uncased")
roberta_gen = pipeline("text-generation", model="roberta-base")
bart_gen = pipeline("text-generation", model="facebook/bart-base")
```

```python
def try_generation(name, model):
    print(f"\n===== {name} =====")
    try:
        output = model(prompt, max_length=30)
        print(output[0]["generated_text"])
    except Exception as e:
        print("Failed:", str(e))


try_generation("BERT", bert_gen)
try_generation("RoBERTa", roberta_gen)
try_generation("BART", bart_gen)
```

```python
masked_sentence = "The goal of Generative AI is to [MASK] new content."

bert_fill = pipeline("fill-mask", model="bert-base-uncased")
roberta_fill = pipeline("fill-mask", model="roberta-base")
bart_fill = pipeline("fill-mask", model="facebook/bart-base")
```

```python
def try_fill(name, model):
    print(f"\n===== {name} =====")
    try:
        preds = model(masked_sentence)
        for p in preds[:5]:
            print(p["token_str"], "->", round(p["score"], 4))
    except Exception as e:
        print("Failed:", str(e))


try_fill("BERT", bert_fill)
try_fill("RoBERTa", roberta_fill)
try_fill("BART", bart_fill)
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

bert_qa = pipeline("question-answering", model="bert-base-uncased")
roberta_qa = pipeline("question-answering", model="roberta-base")
bart_qa = pipeline("question-answering", model="facebook/bart-base")
```

```python
def try_qa(name, model):
    print(f"\n===== {name} =====")
    try:
        answer = model(question=question, context=context)
        print(answer)
    except Exception as e:
        print("Failed:", str(e))


try_qa("BERT", bert_qa)
try_qa("RoBERTa", roberta_qa)
try_qa("BART", bart_qa)
```
