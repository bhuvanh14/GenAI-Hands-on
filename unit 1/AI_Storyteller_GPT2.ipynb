```python
from transformers import pipeline

# Load text generation model
generator = pipeline(
    "text-generation",
    model="gpt2"   # or "distilgpt2" (faster)
)

prompt = input("Start your story: ")

story = generator(
    prompt,
    max_length=150,        # total tokens
    do_sample=True,        # creativity ON
    temperature=0.9,      # randomness level
    top_k=50,             # choose from top 50 words
    top_p=0.95,           # nucleus sampling
    num_return_sequences=1
)

print("\nðŸ“– Story:\n")
print(story[0]["generated_text"])
```
